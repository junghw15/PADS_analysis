PADS 데이터 json --> CSV 변환

```python
import pandas as pd
import csv
import json
```

```python
with open('PADS_data.json') as json_file:
    data = json.load(json_file)
    
data
```

```python
len(data['rows'])
```

```python
PADS_list=[]

for row_number in range(1, len(data['rows'])+1):
    PADS_list.append(list(data['rows']['row{}'.format(row_number)].values()))

print(PADS_list)

df=pd.DataFrame(PADS_list,columns=['PID', 'Detail', 'Pad area', 'Pad width', 'Pad height', 'Mark area', 'Mark area ratio', 'Edge distance(Left)', 'Edge distance(Right)', 'Edge distance(Up)', 'Edge distance(Down)', 'Distance between center(X)', 'Distance between center(Y)', 'result'])

df
```

```python
df.to_csv('PADS_data.csv',index=False)
test=pd.read_csv('PADS_data.csv')
test
```
